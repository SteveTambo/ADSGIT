** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
#creating the figure 

fig = plt.figure()

#create the axes [left, bottom, width , height]

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])


#plotting the figure 

axis1 = ax1.plot(x,z)
axis2 = ax2.plot(x ,y)

#setting the labels for each axes 
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.xlim([20,22])
plt.ylim([30,50])
plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y, 'b',linestyle = 'dashed',linewidth=3)
ax2.plot(x,z, 'r', linewidth=3.5)
plt.show()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2, figsize=(10,2))
ax1.plot(x,y, 'b',linewidth=4)
ax2.plot(x,z, 'r',linestyle = 'dashed',linewidth=2)
plt.show()
```
